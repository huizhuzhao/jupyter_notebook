```python
import numpy as np
import matplotlib.pyplot as plt
from jupyter_notebook.utils import metrics
```

```python
def gen_data(num_data=10):
    x = np.random.randn(num_data)
    y_0 = np.random.randn(num_data)
    y_1 = 0.5 * x
    y_2 = -0.5 * x
    return x, y_0, y_1, y_2

x, y_0, y_1, y_2 = gen_data(100)    
r_0 = metrics.correlation_pearson(x, y_0)
r_1 = metrics.correlation_pearson(x, y_1)
r_2 = metrics.correlation_pearson(x, y_2)


print r_0, r_1, r_2
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True)
ax1.scatter(x, y_0)
ax1.set_title('r=0.04')
ax2.scatter(x, y_1)
ax2.set_title('r=1.0')
ax3.scatter(x, y_2)
ax3.set_title('r=-1.0')
plt.show()
```
